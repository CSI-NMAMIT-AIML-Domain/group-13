```python
import pandas as pd
```

```python
df = pd.read_csv("csi/dataset.csv")

df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt

```

```python


plt.figure(figsize=(10, 6))
plt.scatter(df['Battery Capacity(kWh)'], df['Range(km)'], color='blue')

for i, brand in enumerate(df['Brand Name']):
    plt.text(df['Battery Capacity(kWh)'][i], df['Range(km)'][i], brand, fontsize=8)

plt.title('Battery Capacity vs. Range')
plt.xlabel('Battery Capacity (kWh)')
plt.ylabel('Range (km)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Price(Lh)'], df['TopSpeed(km/h)'], color='green')

for i, brand in enumerate(df['Brand Name']):
    plt.text(df['Price(Lh)'][i], df['TopSpeed(km/h)'][i], brand, fontsize=8)

plt.title('Price vs. Top Speed')
plt.xlabel('Price (Lakh)')
plt.ylabel('Top Speed (km/h)')
plt.grid(True)
plt.show()
```

```python
max_range_car = df.loc[df['Range(km)'].idxmax()]
print(f"The car with the maximum range is {max_range_car['Brand Name']} with a range of {max_range_car['Range(km)']} km.")
```
